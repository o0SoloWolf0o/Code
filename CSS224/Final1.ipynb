1.1 จงเขียนโปรแกรมหรืออัลกอริทึมเพื่อทำการวัดระยะและเทียบลำดับพันธุกรรมจะใช้ภาษาอะไรก็ได้ที่ถนัด (10คะแนน ถ้าไม่สามารถเขียนโค้ดที่รันได้หัก 3 คะแนน)

1.2 จงเขียนโปรแกรมหรืออัลกอริทึมเพื่อทำการสร้างต้นไม้วิวัฒนาการชาติพันธุ์ด้วยขั้นตอนวิธีเชิงละโมบจะใช้ภาษาอะไรก็ได้ที่ถนัด (10 คะแนน ถา้ไม่สามารถเขียนโคด้ ที่รันได้หัก 3 คะแนน)

1.3 จากข้อ 1.1 และ 1.2 จงวิเคราะห์เวลาและพื้นที่ที่ใช้ในการสร้างต้นไม้วิวัฒนาการชาติพันธุ์(5 คะแนน)

1.4 นักศึกษาคิดว่าขั้นตอนวิธีเชิงละโมบดังกล่าวสามารถสร้างต้นไม้ที่ดีที่สุดได้หรือไม่จงวิเคราะห์ (5 คะแนน)
(hint:: เราใช้ขั้น ตอนวิธีเชิงละโมบกับ Matrix Chain Multiplication ได้หรือไม่) 
(hint2 :: สมมุติว่าระยะจาก A-B เท่ากับระยะจาก A-C)

```python
def longest_common_subsequence(str1, str2):
    m = len(str1)
    n = len(str2)

    dp = [[0] * (n + 1) for _ in range(m + 1)]

    for i in range(1, m + 1):
        prev = 0
        for j in range(1, n + 1):
            current = dp[i][j]
            if str1[i - 1] == str2[j - 1]:
                dp[i][j] = dp[i - 1][j - 1] + 1  # แก้ไขตรงนี้เพื่อใช้ค่าจากตาราง dp ที่ถูกต้อง
            else:
                dp[i][j] = max(dp[i - 1][j], dp[i][j - 1])
            prev = current

    lcs = ""
    i, j = m, n
    while i > 0 and j > 0:
        if str1[i - 1] == str2[j - 1]:
            lcs = str1[i - 1] + lcs
            i -= 1
            j -= 1
        elif dp[i - 1][j] > dp[i][j - 1]:
            i -= 1
        else:
            j -= 1

    return lcs, dp[m][n]  # แก้ไขตรงนี้เพื่อส่งค่าความยาวของ LCS ที่ถูกต้อง

def format_output(str1, str2, lcs):
    lcs_str, lcs_length = lcs
    print("ระยะทางระหว่าง '{}' และ '{}' คือ: {}".format(str1, str2, lcs_length))
    print("ลำดับพันธุกรรมที่ใกล้ที่สุด: {}".format(lcs_str))

string1 = "TGCTACTCCACACAC"
string2 = "TAGTACTCCACACAC"
string3 = "TGCTACTGCAGACAC"
string4 = "TACTACTGCACCCAC"
string5 = "TACCACTCCACACCC"

lcs = longest_common_subsequence(string1, string2)
format_output(string1, string2, lcs)
```

```python
def create_evolutionary_tree(string1, string2):
    if len(string1) != len(string2):
        return None
    
    result = ""
    for char1, char2 in zip(string1, string2):
        if char1 == char2:
            result += char1
        else:
            result += "-"
    
    return result

# ตัวอย่างการใช้งาน
string1 = "TGCTACTCCACACAC"
string2 = "TAGTACTCCACACAC"
result = create_evolutionary_tree(string1, string2)
print("A/B: ",result)
```

1.3 จากข้อ 1.1 และ 1.2 จงวิเคราะห์เวลาและพื้นที่ที่ใช้ในการสร้างต้นไม้วิวัฒนาการชาติพันธุ์(5 คะแนน)

1.1 มี Time Complexity เท่ากับ O(mn) และ Space Complexity เท่ากับ O(mn) โดยที่ m และ n คือจำนวนลำดับพันธุกรรมที่มีอยู่ในต้นไม้

1.2 มี Time Complexity เท่ากับ O(n) และ Space Complexity เท่ากับ O(n)

1.4 นักศึกษาคิดว่าขั้นตอนวิธีเชิงละโมบ(Greedy)ดังกล่าวสามารถสร้างต้นไม้ที่ดีที่สุดได้หรือไม่จงวิเคราะห์ (5 คะแนน)
(hint:: เราใช้ขั้น ตอนวิธีเชิงละโมบกับ Matrix Chain Multiplication ได้หรือไม่) 
(hint2 :: สมมุติว่าระยะจาก A-B เท่ากับระยะจาก A-C)

ขั้นตอนวิธีเชิงละโมบ (Greedy) ที่ใช้ในฟังก์ชัน create_evolutionary_tree ในที่นี้ไม่สามารถสร้างต้นไม้ที่ดีที่สุดได้เสมอ โดยเฉพาะเมื่อสองสตริงที่กำหนดมีความยาวต่างกัน

วิธีการทำงานของฟังก์ชัน create_evolutionary_tree คือการเปรียบเทียบสองสตริงตามตำแหน่งและสร้างสตริงผลลัพธ์ที่มีเครื่องหมาย "-" ในตำแหน่งที่สองสตริงไม่เหมือนกัน ดังนั้นผลลัพธ์ที่ได้จากฟังก์ชันนี้จะเป็นสตริงที่แสดงความแตกต่างของสองสตริงที่กำหนด ด้วยเครื่องหมาย "-" ในตำแหน่งที่ไม่ตรงกัน

การใช้วิธีการเชิงละโมบกับ Matrix Chain Multiplication (MCM) เป็นเรื่องที่แตกต่างจากการใช้วิธีเชิงละโมบกับการสร้างต้นไม้ทางวิวัฒนาการ (evolutionary tree) เพราะ MCM เป็นปัญหาทางคณิตศาสตร์ที่มีลักษณะเป็นแบบปัญหาการแบ่งแยก (divide and conquer) และมีลักษณะเชิงคำนวณที่เหมาะสมกับการใช้เทคนิคเชิงละโมบ เช่นการใช้ Dynamic Programming เพื่อคำนวณค่าน้อยที่สุดของการคูณสตริงหลายๆ สตริง (matrix) ที่กำหนดให้

นักวิทยาศาสตร์คอมพิวเตอร์ใช้วิธีการเชิงละโมบกับ MCM เพื่อลดปัญหาการคำนวณที่เกิดขึ้นใน MCM ด้วยการคำนวณผลลัพธ์ของแต่ละสตริงที่เกี่ยวข้องแยกกันเป็นแทนที่จากนั้นรวมผลลัพธ์เหล่านั้นเข้าด้วยกันเพื่อได้ผลลัพธ์สุดท้าย

ดังนั้น การใช้วิธีการเชิงละโมบกับ MCM มีลักษณะและขั้นตอนการทำงานที่แตกต่างจากการสร้างต้นไม้ทางวิวัฒนาการใน create_evolutionary_tree
