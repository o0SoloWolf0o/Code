```python
def calculate_price(piece_counts):
    piece_values = {
        'Q': 9,
        'K': 4,
        'B': 3,
        'N': 3,
        'P': 1,
        'R': 5
    }
    
    total_cost = 0
    for piece, count in piece_counts.items():
        total_cost += piece_values[piece] * count
    
    return total_cost


input_string = input().strip()
piece_counts = {}

for piece in input_string:
    if piece in piece_counts:
        piece_counts[piece] += 1
    else:
        piece_counts[piece] = 1

if 'K' in piece_counts:
    total_cost = calculate_price(piece_counts) * 200
else:
    total_cost = calculate_price(piece_counts) * 100

print(total_cost)
```

```python
n = int(input())
# ผลรวมอนุกรม
# (1 + 2 + 3 + ... + N)
S1 = n * (n + 1) // 2
# ผลรวมของจำนวนเต็มบวก N จำนวนแรกที่ถูกยกกำลังสอง
# (1**2 + 2**2 + 3**2 + ... + N**2)
S2 = n * (n + 1) * (2 * n + 1) // 6
# ผลรวมของจำนวนเต็มบวก N จำนวนแรกที่ถูกยกกำลังสาม
# (1**3 + 2**3 + 3**3 + ... + N**3)
S3 = (n * (n + 1) // 2) ** 2

Total_Sum = S1 + S2 + S3
print(Total_Sum)
```
