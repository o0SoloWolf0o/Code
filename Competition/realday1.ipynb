```python
# You will be given a string of digits to add up. Start adding them up, but if at any time one of the digits is 0, reset the sum back to 0.
# Input
# Line 1: An integer N telling the number of digits in the string.
# Line 2: A string S of N digits.
# Output
# The sum, given as an integer.
# Constraints
# 1 ≤ N ≤ 1000

# Example
# Input
# 5
# 12345
# Output
# 15


N = int(input())
S = input()
sum = 0
for i in range(N):
    if S[i] == '0':
        sum = 0
    else:
        sum += int(S[i])
print(sum)
```

```python
import math

input_string = input()

ascii_sum = sum(ord(char) for char in input_string)

result = math.isqrt(ascii_sum)

print(result)
```

```python
input_string = input().split()
count = 0

emoticons = ["^_^", "=)"]

for word in input_string:
    if word in emoticons:
        count += 1
    else:
        count -= 1

print(count)
```

```python
def max_div_min(A, B, C):
    maximum = max(A, B, C)
    minimum = min(A, B, C)
    if minimum == 0:
        minimum = 1
    result = maximum / minimum
    return result
a,b,c = int(input()),int(input()),int(input())
print(int(max_div_min(a,b,c)))
```
