```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('image.png',0)
# Roberts
kernel = np.array(([-1,0],
                   [0,1]), dtype='int')

output = cv2.filter2D(img, -1, kernel)

numpy_horizontal = np.hstack((img, output))

plt.imshow(numpy_horizontal, cmap='gray')
```
